# we apply the interpolation strategy for night calls and night minutes as well

```python
# we interpolate night calls here
df['Total night calls.interpolate'] = df['Total night calls'].interpolate(method='nearest', limit_direction='forward')
df['Total night calls'].hist()
df['Total night calls'].mean()
```

```python
# we interpolate night minutes here 
df['Total night minutes.interpolate'] = df['Total night minutes'].interpolate(method='nearest', limit_direction='forward')
df['Total night minutes'].hist()
df['Total night minutes'].mean()
```

```python
# we shall check if the interpolation was successful in removing the null values
df.isnull().sum()
```

# over here we shall use mean imputer for filling in the null values for Total day minutes and Total eve minutes
# and we shall use mode for Voice mail plan and international plan as there is no other way to identify missing non-numeric 
# data


```python
from sklearn.impute import SimpleImputer
```

```python
df_mean = df.copy(deep=True)
dfm = SimpleImputer(strategy='mean')
```

```python
df_mean[['Total day minutes', 'Total eve minutes']] = dfm.fit_transform(df_mean[['Total day minutes', 'Total eve minutes']])
```

```python
# now we shall test if the above code worked
df_mean.isnull().sum()
```

```python
df.isnull().sum()
```
